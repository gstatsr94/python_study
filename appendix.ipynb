# 부록 : 프로그래밍 능력향상을 위한 조언

이 Notebook은 ‘알기 쉬운 파이썬 제4판’을 읽은 독자 여러분에게, 그 다음은 무엇을 해야 하는지 책에서는 적지 못한 것을 알려드리려고 합니다. 프로그래밍을 향상시키기 위한 조언을 곰곰이 적어 보았습니다.

## 1. 목표를 설정하자

먼저, 파이썬을 사용하여 무엇을 하고 싶은지 목표를 명확히 해야 합니다. “Web 응용 프로그램을 만들고 싶다”, “Raspberry Pi에서 IoT 장치를 만들고 싶다”, “인공지능을 활용하고 싶다”, “데이터 분석을 하고 싶다” 등 가능한 한 구체적인 목표를 세워야 합니다.

“세상을 바꾸고 싶다”, “돈을 벌고 싶다”, “이성에게 인기를 얻고 싶다”, “어쨌든 빅(Big)이 되고 싶다”라는 추상적인 목표를 세워서는 안 됩니다. 번뇌는 지적 성장에 도움이 되지 않습니다.

## 2. 간단한 것부터 시작하자

목표를 세웠으면 목표를 향해서 느려도 좋으니 천천히 출발합시다.

프로그래밍을 숙달하는 비결은 매일 프로그램을 작성하는 것입니다. 우선 간단한 프로그램부터 작성을 시작합니다.

먼저, 어떤 프로그램을 만들고 싶은지를 결정합니다. 그 다음 프로그램을 만드는 방법을 머리로 생각해 봅니다. 사고 실험을 해보는 것입니다.

그때 처리의 실현 방법이 상상할 수 없을 정도로 어려운 부분이 전혀 없는 단순함이 좋다고 생각합니다. 어려움이 있더라도 기껏해야 한두 가지 정도의 단순함이 좋습니다.

상상 밖에 있는 것은 실현 불가능합니다.

## 3. 지루한 작업을 자동화하자

그러면 구체적으로 어떤 프로그램을 만들면 좋을까요. 제일 먼저 추천하는 방법은 일상적인 작업을 자동화하는 프로그램입니다.

일상의 업무나 생활에서 매일 실행하는 것이라면 뭐든지 좋습니다. 데이터 집계, 그날의 날씨를 이메일로 보내는 등 수동으로 실행하면 몇 분 소요될 같은 작업이 반드시 있어야 합니다. 그런 작업을 파이썬을 사용하여 자동화하는 것입니다. 또는 자신을 위한 프로그램이 아니라 가족과 친구들을 위한 프로그램을 만들어도 좋습니다.
한 번 만든 프로그램은 조금씩 개량해 갑니다. 만든 프로그램에 만족하면 다른 작업을 자동화하는 새로운 프로그램을 만들어 봅시다. 이렇게 하면 매일 프로그램을 작성하는 습관을 익힐 수 있습니다.

그리고 여기가 중요합니다만, 만약 업무 작업을 파이썬으로 자동화하면 그것을 상사나 동료에게 이야기해서는 안됩니다. 작업을 자동화하고 남은 시간은 프로그램의 공부에 활용합니다. 그렇게 공부 시간을 늘려 가면, 프로그래밍 능력이 향상됩니다. 프로그래밍 능력이 향상되면 더 복잡한 작업을 자동화할 수 있게 되어 더 많은 시간을 공부하는 데 투자할 수 있게 됩니다.

편하게 하기 위해 노력해야 합니다.

## 4. “바퀴의 재발명”은 나쁘지 않다

잘 알려진 도구를 미처 알지 못하고 스스로 만들어 버리는 것을 “바퀴의 재발명”이라고 합니다. 쓸데없는 비유로 자주 사용되는 단어입니다만, 바퀴의 재발명은 전혀 소용이 없습니다. 왜냐하면 바퀴를 다시 발명하는 것으로, 바퀴 만드는 방법을 배우기 때문입니다.

프로그래밍에서는 이미 있는 프로그램이나 라이브러리 등을 직접 스스로 만드는 것을 바퀴의 발명이라고 합니다. 파이썬에 한정하면, 예를 들어 표준 라이브러리에 있는 함수와 클래스, 내장된 메쏘드의 기능을 일부러 스스로 만들어 보는 것 등이 이에 해당합니다.

내장된 메쏘드에 같은 기능은 처리가 간단합니다. 특히 초보자 가운데는 그런 기능을 직접 구현 보는 것은 매우 좋은 공부가 됩니다.

바퀴를 다시 발명한 결과, 쓸데없는 시간을 사용하여 버리거나, 프로그램의 전망이 나빠져 버릴지도 모릅니다. 그러나 스스로 재발견한 코드를 기존의 기능에서 대체하여 프로그램의 질을 높이는 것의 중요성을 몸으로 알게 될 것입니다.

고통이 피와 살이 됩니다.

## 5. 도전하는 마음을 잃지 말자

“쉬운 것부터 시작한다.”라는 것은 낮은 계단부터 차근차근 확실히 올라가는 것과 같습니다. 이 방법은 편하고 확실하지만, 높은 곳으로 올라가는데 시간이 걸리고, 계단을 올라가는데 지루해질 수 있다는 단점도 있습니다.

프로그래밍에 익숙해지면 조금 어려운 번째 과제를 설정하는 것도 재미있습니다. 계단에 오르면서 비틀거림에 대한 내성이 생기면 도전하는 마음이 저절로 싹틀지도 모릅니다.

그럴 때는 벽처럼 우뚝 솟은 높은 계단을 기어오르는 것처럼 어려운 과제를 설정해 봐도 좋을 것입니다. 높은 벽에 막혀 돌아갈 때는 의식하지 않고 낮은 계단으로 돌아가면 좋습니다.

높은 벽을 극복하면 기쁨도 큽니다.

## 6. 코드를 작성하기 전에 먼저 조사해 보자

프로그래밍 학습이 진전되면 조금씩 고급 과제를 수행하게 됩니다. 프로그램을 만들 때 모르는 것이 나오면, 우선 인터넷 등을 통해 알아보도록 합니다.

파이썬의 경우는 과제를 한번에 해결해 주는 모듈을 찾을 가능성도 매우 큽니다. 우선 표준 라이브러리를 검색한 다음, 타사 라이브러리에 대해서도 조사해 봅시다.

조사 과정에서 데이터 구조 및 알고리즘 등, 프로그래밍에 대해 좀 더 추상적인 지식을 배울 수 있는 기회가 늘어날지도 모릅니다. 그렇게 되면 여러분은 초보자를 뛰어넘어 중급자가 되고 있는 증거입니다.



## 7. 단순히 복사 & 붙여넣기 하는 것을 피하고, 코드를 이해하자

인터넷을 살펴보면 프로덕션 코드 조각이 많이 발견됩니다. 블로그와 코드 공유 서비스에서 많은 프로그램이 공개되어 있는 현대는 프로그램을 배울 수 있는 아주 좋은 시대라고 생각합니다.

인터넷에서 찾은 코드는 가능한 내용을 이해한 후 사용해야 합니다. 그대로 복사 & 붙여넣기 하는 것이 아니라, 코드를 읽고 프로그램의 움직임을 파악하고 사용하도록 합니다. 코드를 자기 방식으로 고쳐 쓰면 프로그래밍 학습에도 매우 효과적입니다.

코드의 행간을 읽는 것입니다.

## 8. 고품질의 코드를 읽자

프로그램은 만드는 것 외에도 읽기를 통해서 배울 수 있습니다. 프로그램을 읽는 것을 “코드 읽기”라고 합니다. 코드 리딩도 프로그래밍 학습에 효과적인 방법입니다.

코드 리딩을 할 때는 가능한 양질의 코드를 읽어야 합니다. 유명한 프레임워크와 라이브러리, 파이썬 표준 라이브러리에 포함되어 있는 것들 중 파이썬으로 작성된 모듈 등을 검색하면 고품질의 코드를 읽을 수 있습니다. 이러한 코드의 대부분은 적절한 주석이 붙어 있어서 처리가 깔끔하고, 매우 읽기 쉽습니다.

읽기 쉬운 코드야말로 좋은 코드입니다.

## 9. 지식을 옆으로 넓히자

목표로 하는 분야의 프로그램 제작에 숙달되면 관점을 바꾸어 봅시다. 시야를 약간 옮겨 익숙한 분야에 가까운 다른 분야를 선택하여 프로그램을 만들어 봅시다.

한마디로 프로그래밍이란, 분야가 다르면 다른 종류의 지식을 요구합니다. 비슷한 분야의 프로그램을 만들어 봄으로써 지식의 폭을 넓힐 수 있으며, 더 많은 문제를 프로그래밍 능력으로 해결할 수 있게 됩니다.

좋은 지식은 T 자형의 분포를 나타냅니다.

## 10. 책으로 학습하는 것이 효율적이다

프로그래밍 공부는 인터넷으로 충분하다는 사람이 종종 있습니다. 이 이야기는 반은 맞고, 반은 잘못된 것입니다.

인터넷에는 프로그래밍에 대해 상당한 양의 정보가 공개되어 있습니다. 그런 정보는 대부분 작성자 본인의 문제를 해결하는 과정을 공개하는 것에 지나지 않습니다. 따라서 취급 범위가 좁고 흩어져있어 통합되어 있지 않습니다.

독자의 생각, 특정 분야에 대해 배워야 하는 에센스를 필요한 순서대로 정리한 책을 사용합시다. 그런 책으로 배우면 보다 짧은 시간에 필요한 지식을 얻을 수 있습니다. 여러분의 시급을 과소 평가하여 8,000원이라 합시다. 인터넷에서 단편적인 정보를 공부하느라 4시간을 소비하는 것보다는 20,000원을 들여 책을 사서 공부하는 것이 훨씬 더 유익합니다.

싼 것만을 추구하는 요리사는 가난한 길로 가게 됩니다.

## 11. 도구에 신경을 쓰자

힘든 고통을 웃는 얼굴로 짊어져서는 안됩니다. 귀찮게 생각되는 것이 있으면 최대한의 노력으로 쉽게 처리하는 방법을 생각하세요.

프로그램을 만들기 위해서는 다양한 도구를 사용합니다. 에디터(editor)는 누구나 사용하는 도구입니다. 코드를 작성하는 시간을 절감해 주는 손에 익숙한 에디터를 선택하여 사용합니다. 많은 도구를 사용해 손에 익숙한 에디터를 사용하면 생산성이 크게 향상됩니다. 편집기에 준하는 IDE도 마찬가지입니다.

중급자가 되면 테스트 및 CI 도구를 사용하게 될 것입니다. 많은 도구를 테스트해 본 다음, 사용하기 쉬운 도구를 스스로 찾을 수 있도록 합니다.

귀찮은 일에 질색하는 사람이야말로 훌륭한 프로그래머가 될 수 있습니다.

## 12. 멘토를 찾아라

프로그래밍의 스승을 찾는 것은 의외로 어렵습니다. 스승이 될 수 있는 사람이 절대 부족하기 때문입니다.

스승이 될 수 있는 사람은 대부분 바쁘며, 통상 접촉하는 것이 어렵습니다. 다만, 지금은 인터넷이 있습니다. SNS나 GitHub도 있습니다. 이런 시대에 멘토와 연결하기 위해서는 어떤 전략을 세우면 좋을지 생각해 봅시다.

먼저, ‘이 사람이다’라고 생각되는 사람을 찾아보세요. 인터넷에서 그 사람과 연결하는 방법을 찾아 보고, 다음 질문 방법을 잘 생각합니다. 상대를 존중하고 상대의 시간을 최대한 없애지 않는 질문 방법을 생각합니다. 질문을 하기 전에 문제를 머리에서 되새겨보고, 자신이 모르는 것이 사실 뭔가를 이해하고 질문을 작성합니다.

멘토가 될 수 있는 좋은 사람은 정중하고 핵심을 찌르는 질문은 내버려두지 않습니다.

## 13. 영어를 공부하자

많은 새로운 기술이 외국, 특히 영어권에서 옵니다. 이러한 기술의 설명이 번역되어 나올 무렵에는 그 분야는 이미 치열한 전쟁터로 변해있으며 무례한 자들에 의해 죽음의 땅이 되어 버립니다.

신선하고 경쟁력 있는 기술을 재빨리 자신의 것으로 만들기 위해서는 최소한 영어를 읽을 수 있어야 합니다. 영어를 쓸 수 있다면, 기본 정보에 접속할 수 정보를 보낼 수 있습니다. 영어를 말할 수 있다면, 외국에 가서 생생한 정보를 접하거나 기술의 성장점의 분위기를 느낄 수 있는 큰 동기를 얻을 수 있습니다.

영어로 문학 작품을 쓰고 싶다든가, 할리우드와 브로드웨이에서 배우로서 성공하겠다는 것이 아니라면 문법과 발음에 집착할 필요는 없습니다. IT 세계에서 사용되고 있는 영어는 원래 오해를 두려워하지 말고 말하면 “엉망”입니다. 이 분야는 유럽과 아시아 사람들이 많이 일하고 있습니다. 영어를 모국어로 하는 사람들도 그런 상황에 익숙해져 있기 때문에 엉성한 영어에 관용을 보입니다. 그들이 이해하지 못 하면 물어옵니다. 그리고 마찬가지로 이해가 안 되면 물어보면 됩니다.

영어를 사용하는데 있어서 가장 필요한 것은 문법이 아니라 전달하고자 하는 의지입니다.
